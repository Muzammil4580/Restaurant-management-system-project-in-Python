```python
import pandas
import csv
print("\n\nWelcome to Restaurant Management System")

def displayMain():
    print("For Rate list Press 1")
    print("For Sales Details Press 2")
    print("For Employes Details Press 3")
    print("For Exit Press 4")
    a = int(input("Enter the number in menu you want To proceed"))
    if a==1:
        print("\n\nWelcome to Ratelist Menu")
        rate()
    elif a==2:
        print("\n\nWelcome to Sales Details Menu")
        sales()
    elif a==3:
        print("\n\nWelcome to Employe Details Menu")
        employe()
    else:
        quit()
    return
####
################

def rate():
    while 1:
        print("\n\n1.Show\n2.Update\n3.Main Menu\n")
        b = int(input("Enter the number from Menu to Proceed"))
        if b==1:
            milk=pandas.read_csv('rate.csv')
            print(milk)
        elif b==2:
            n = int(input("Enter the Serial number in which replacement is made: "))
            r = str(input("Enter the new thing"))
            v = n-1
            df = pandas.read_csv('rate.csv')
            df.loc [v, 'Rate'] = r
            df.to_csv('rate.csv', index = False)
            print(df)
       
        elif b==3:
            displayMain()
##############################################


#===================================================

def sales():
    while 1:
        print("\n\n1.Insert\n2.Show\n3.Main Menu\n")
        c = int(input("Enter the number from Menu to Proceed"))
        if c==1:
            from csv import writer
            am = str(input("Enter Date"))
            bm = str(input("Enter Item Name"))
            cm = int(input("Enter Item Quantity"))
            dm = int(input("Enter Total Bill"))
            list_data = [am, bm, cm, dm]
            with open('Sales.csv', 'a', newline='') as f_object:
                writer_object= writer(f_object)
                writer_object.writerow(list_data)
                f_object.close()
        elif c==2:
            milk=pandas.read_csv('Sales.csv')
            print(milk)
        elif c==3:
            displayMain()
##############################################


#===================================================



#===================================================
def employe():
    while 1:
        print("\n\n1.Insert\n2.Show\n3.Update\n4.Delete\n5.Main Menu\n")
        g= int(input("Enter the number from Menu to Proceed"))
        if g==1:
            from csv import writer
            am = str(input("Enter ID No: "))
            bm = str(input("Enter Employe Name"))
            cm = str(input("Enter Employe Status"))
            dm = int(input("Enter Salary"))
            list_data = [am, bm, cm, dm]
            with open('Employe.csv', 'a', newline='') as f_object:
                writer_object= writer(f_object)
                writer_object.writerow(list_data)
                f_object.close()
        elif g==2:
            milk=pandas.read_csv('Employe.csv')
            print(milk)
        elif g==3:
            n = int(input("Enter the number of row in which replacement is made: "))
            N = str(input("Enter the name of column in which you want replacement"))
            r = str(input("Enter the new thing"))
            v = n-1
            df = pandas.read_csv('Employe.csv')
            df.loc [v, N] = r
            df.to_csv('Liab.csv', index = False)
            print(df)
        elif g==4:
            n = int(input("Enter the ID of Employe Whose Data you want to delete:"))
            v = pandas.read_csv('Employe.CSV')
            d = v[: 4]
            d.set_index('Employe ID', inplace=True)
            d = d.drop(n)
            print(d)
        elif g==5:
            displayMain()
##############################################


displayMain()
```
